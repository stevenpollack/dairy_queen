```python
import requests, re
```

```python
dairy_queen_url = 'https://dairy-queen.herokuapp.com/double-dips'
params = {
    'location': 'Berlin',
    'days_from_now': 1
}
```

```python
r = requests.get(dairy_queen_url, params=params)
```

```python
def filter_sony_center(x):
    name = x.get('name')
    info = x.get('info')
    if re.match('CineStar', name) and re.match('.*Sony.*', info):
        return True
    else:
        return False
```

```python
sony_center = next(filter(filter_sony_center, r.json()))
```

```python
for double_dip in sony_center.get('doubleDips'):
    # double_dip is an array of movies
    print(' \u2192 '.join(["%s %s-%s" % (movie.get('movie'), movie.get('startTime'), movie.get('endTime')) for movie in double_dip]))
    print('\n')
```
